```python
import sys
import snap
import numpy as np

import matplotlib.pyplot as plt
```

## Question 1

```python
head_name = "epinions"
G = snap.LoadEdgeList(snap.PNGraph, "soc-Epinions1.txt", 0, 1)
snap.PlotInDegDistr(G, head_name, "G_wiki In Degree")
indegree_o = np.loadtxt("inDeg."+head_name+".tab")
snap.PlotOutDegDistr(G, head_name, "G_wiki Out Degree")
outdegree_o = np.loadtxt("outDeg."+head_name+".tab")

# filter zero entities
filtered = outdegree_o[:,1] > 1 
outdegree = outdegree_o[filtered, :]

# filter zero entities
filtered = indegree_o[:,1] > 1 
indegree = indegree_o[filtered, :]

z = np.polyfit(np.log10(outdegree[1:,0]), np.log10(outdegree[1:,1]), 1)
a = z[0]; b = z[1];

estimated_degree = np.power(10,b)*np.power(outdegree[1:,0], a)
```

```python
plt.figure(figsize=(12,3))
plt.subplot(121)
plt.loglog(outdegree_o[:,0], outdegree_o[:,1])
plt.loglog(outdegree[1:,0], estimated_degree)
plt.title(head_name +":Log-Log plot of the out-degree")
plt.subplot(122)
plt.loglog(indegree_o[:,0], indegree_o[:,1])
plt.title(head_name+":Log-Log plot of the in-degree")
```

# Question 3

```python
Graph1 = snap.LoadEdgeList(snap.PUNGraph, 'graph1.txt', 0, 1)
Graph2 = snap.LoadEdgeList(snap.PUNGraph, 'graph2.txt', 0, 1)

snap.PlotInDegDistr(Graph1, "graph1", "G_wiki In Degree")
indegree_o_1 = np.loadtxt("inDeg."+"graph1"+".tab")

snap.PlotInDegDistr(Graph2, "graph2", "G_wiki In Degree")
indegree_o_2 = np.loadtxt("inDeg."+"graph2"+".tab")

# filter zero entities
filtered = indegree_o_1[:,1] > 1 
indegree_1 = indegree_o_1[filtered, :]

# filter zero entities
filtered = indegree_o_2[:,1] > 4
indegree_2 = indegree_o_2[filtered, :]

z = np.polyfit(np.log10(indegree_1[1:,0]), np.log10(indegree_1[1:,1]), 1)
a = z[0]; b = z[1];

estimated_degree_1 = np.power(10,b)*np.power(indegree_1[1:,0], a)

z = np.polyfit(np.log10(indegree_2[1:,0]), np.log10(indegree_2[1:,1]), 1)
a = z[0]; b = z[1];

estimated_degree_2 = np.power(10,b)*np.power(indegree_2[1:,0], a)
```

```python
plt.figure(figsize=(12,3))
plt.subplot(121)
plt.loglog(indegree_o_1[:,0], indegree_o_1[:,1])
plt.loglog(indegree_1[1:,0], estimated_degree_1)
plt.title("graph1 :Log-Log plot of the degree")
plt.xlabel("degree")
plt.ylabel("#nodes")
plt.subplot(122)
plt.loglog(indegree_o_2[:,0], indegree_o_2[:,1])
plt.loglog(indegree_2[1:,0], estimated_degree_2)
plt.title("graph2 :Log-Log plot of the degree")
plt.xlabel("degree")
plt.ylabel("#nodes")
```
