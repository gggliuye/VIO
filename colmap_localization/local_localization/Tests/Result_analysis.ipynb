```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
import glob
```

```python
image_name = "debug.png"
log_name = "log.txt"

result_folders_1 = ["res1", "res3", "res5"]
result_folders_2 = ["res2", "res4", "res6"]
```

### Load runtime results

runtimefile << pCurrentFrame->CountMatches() << " " << time_sift_d << " " << time_pose_d
            << " " << time_process_t << " " << noise_t.norm() << " " << distance_e;

```python
debug_images_1 = []
log_results_1 = []
noise_levels_1 = []
success_rates_1 = []
average_time_1 = []

for folder in result_folders_1:
    ax = np.loadtxt(folder + "/" + log_name)
    
    noise_t = np.mean(ax[:,4])
    noise_levels_1.append(noise_t)
    
    time_t = np.mean(ax[:,3])
    average_time_1.append(time_t)
    
    success_rates_1.append(ax.shape[0]/75)

    image_file = folder + "/" + image_name
    debug_images_1.append(image_file)
    log_results_1.append(ax)
#log_results_1 = np.asarray(log_results_1)
```

```python
debug_images_2 = []
log_results_2 = []
noise_levels_2 = []
success_rates_2 = []
average_time_2 = []

for folder in result_folders_2:
    ax = np.loadtxt(folder + "/" + log_name)
    
    noise_t = np.mean(ax[:,4])
    noise_levels_2.append(noise_t)
    
    time_t = np.mean(ax[:,3])
    average_time_2.append(time_t)
    
    success_rates_2.append(ax.shape[0]/75)

    image_file = folder + "/" + image_name
    debug_images_2.append(image_file)
    log_results_2.append(ax)

#log_results_2 = np.asarray(log_results_2)
```

```python
plt.figure(figsize=(15,7))
n = len(log_results_2)
plt.suptitle('Derivation histograms for Local Search Match and FLANN Match')#, verticalalignment='bottom')
for i in range(n):
    range_plot = (0, 1.5)
    res_1 = log_results_1[i][:,5]
    res_1[res_1 > range_plot[1]] = range_plot[1]
    plt.subplot(2,n,(i+1))
    plt.hist(res_1, bins=15, range = range_plot)
    plt.title("Local Search Match with noise level " + str(noise_levels_1[i])[0:4])
    res_2 = log_results_2[i][:,5]
    res_2[res_2 > range_plot[1]] = range_plot[1]
    plt.subplot(2,n,(i+1+n))
    plt.hist(res_2, bins=15, range = range_plot)
    plt.title("FLANN Match with noise level " + str(noise_levels_2[i])[0:4])
```

```python
print("Average time for Local Search Match is", np.mean(average_time_1))
print("Average time for FLANN Match is", np.mean(average_time_2))
print("Average success rate for Local Search Match is", np.mean(success_rates_1))
print("Average success rate for FLANN Match is", np.mean(success_rates_2))
```
