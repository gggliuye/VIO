```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def f(x, y):
    return (x * y - 1)*(x * y - 1)*2
```

```python
def linearlized(x,y,x0,y0):
    jx = 2 * (x0*y0-1) * y0
    jy = 2 * (x0*y0-1) * x0
    dx = x - x0
    dy = y - y0
    return f(x0,y0) + (jx * dx + jy* dy) + jx*jx*dx*dx + 2*jx*jy*dx*dy + jy*jy*dy*dy
```

```python
colmap = 'coolwarm'# 'twilight_shifted'
```

```python
x = np.linspace(0, 2, 50)
y = np.linspace(0, 2, 50)

X, Y = np.meshgrid(x, y)

plt.figure(figsize=(50,10))

Z = f(X, Y)
plt.subplot(141)
plt.title("E1 + E2", fontsize=40)
plt.contourf(X, Y, Z, 40, cmap=colmap)
plt.text(1.4, 0.5, 'O', fontsize=40)
plt.text(0.6, 1.2, 'X', fontsize=40)
plt.axis('off')

Z1 = linearlized(X,Y, 1.4, 0.5)
plt.subplot(142)
plt.title("E1' = E1 linearlized at (1.4,0.5)", fontsize=40)
plt.contourf(X, Y, Z1, 40, cmap=colmap)
plt.text(1.4, 0.5, 'O', fontsize=40)
plt.axis('off')

Z2 = linearlized(X,Y, 0.6, 1.2)
plt.subplot(143)
plt.title("E2' = E2 linearlized at (0.6,1.2)", fontsize=40)
plt.contourf(X, Y, Z2, 40, cmap=colmap)
plt.text(0.6, 1.2, 'X', fontsize=40)
plt.axis('off')

Z3 = Z1 + Z2
plt.subplot(144)
plt.title("E1' +  E2'", fontsize=40)
plt.contourf(X, Y, Z3, 40, cmap=colmap)
plt.axis('off')
```
